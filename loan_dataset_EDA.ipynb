# Data Pre-Processing

### a) Missing values handling

```python
loan_df.isnull().sum()
```

```python
loan_df.dtypes
```

```python
loan_df['paid_off_time'] = loan_df['paid_off_time'].fillna(loan_df['due_date'])
```

#### If the payment time is not recorded, assume the loan was paid on the due date.Because the due date is the last date of repayment .

```python
loan_df.isnull().sum()
```

```python
loan_df.loc[(loan_df['loan_status'] == 'PAIDOFF') & (loan_df['past_due_days'].isnull()), 'past_due_days'] = 0
loan_df['past_due_days'].fillna(-1, inplace=True)      #0 paid  -1 not paid yet
```

#### For loans with status = "PAIDOFF" and missing past_due_days, it sets the value to 0 → meaning no delay.

#### For all other missing values, it fills them with -1 → meaning data not available or loan not paid yet

```python
loan_df.isnull().sum()
```

### b) Outlier detection and handling 

```python
colss=['Principal','terms','past_due_days','age']
for col in colss:
    plt.figure(figsize=(4, 5))
    plt.boxplot(loan_df[col])
    plt.title(f'Boxplot of {col}')
    plt.ylabel(col)
 
```

```python
colss = ['Principal','past_due_days','age']

for col in colss:
    Q1 = loan_df[col].quantile(0.25)
    Q3 = loan_df[col].quantile(0.75)
    IQR = Q3 - Q1
    lower_cap = Q1 - 1.5 * IQR
    upper_cap = Q3 + 1.5 * IQR
    loan_df[colss]=loan_df[colss].clip(lower_cap,upper_cap)
```

```python
num_cols = ['Principal','terms','past_due_days','age']

for col in num_cols:
    plt.figure()
    loan_df.boxplot(column=[col])
    plt.title(f"after Capping: {col}")
    
    
```
